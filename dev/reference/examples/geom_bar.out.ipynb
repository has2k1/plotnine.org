# Bar chart

````python

from plotnine import ggplot, aes, geom_bar, labs, coord_flip, theme_classic
from plotnine.data import mpg
````

*geom_bar() makes the height of the bar proportional to the number of
cases in each group*

````python
mpg.head()
````

### Bar Plot with Counts

````python
# Gallery, distributions
ggplot(mpg) + geom_bar(aes(x="class"))
````

### Sorting the bars

We can use the internal function `reorder` to sort the bars depending on
the size of each group. In this case we create a variable `class2` with
values equal to those of `class` so because `class` is a reserved name
and:

``` python
reorder(class, class, len)
```

is not a valid python expression, while:

``` python
reorder(class2, class2, len)
```

is a valid python expression.

````python
mpg["class2"] = mpg["class"]

(
    ggplot(mpg)
    + geom_bar(aes(x="reorder(class2, class2, len)"))
    + labs(x="class")
)
````

And in descending order

````python
(
    ggplot(mpg)
    + geom_bar(
        aes(x="reorder(class2, class2, len, ascending=False)")
    )
    + labs(x="class")
)
````

````python
(
    ggplot(mpg)
    + geom_bar(aes(x="class", fill="drv"))
)
````

### Horizontal Bar Plot with Counts

````python
# Gallery, distributions
(
    ggplot(mpg) 
    + geom_bar(aes(x="class", fill="drv"))
    + coord_flip()
    + theme_classic()
)
````
